```python
#change the format of a few columns 
autos["price"]=autos.loc[:,"price"].str.replace("$","").str.replace(",","").astype("int")
autos["odometer"]=autos.loc[:,"odometer"].str.replace(",","").str.replace("km","").astype("int")
autos.rename({"odometer":"odometer_km"},axis=1,inplace=True)
```

```python
#investigate the stats of the integer and float columns
autos.describe()
```

```python
#investigate unique values and filter out some rows with bad data 
autos["price"].unique().shape
autos["price"].value_counts().sort_index(ascending=True)
autos=autos[~autos["price"].between(0,100)]
autos=autos[~autos["price"].between(10000000,100000000)]
autos["odometer_km"].unique().shape
autos["odometer_km"].value_counts().sort_index(ascending=True)
autos=autos.drop(["nr_of_pictures"],axis=1)
```

In the previous step, I have removed rows with "price" between 0 and 100 and over 90,000,000 german dollars. The odometer_km values seem to be within a good range. We have now deleted 1897 rows with poor data. I also deleted the nr_of_photos column because it only contain 0s.

Next Step #5: Split the odometer_km into groups, and use aggregation to see if average prices follow any patterns based on mileage. 

```python
autos["odometer_km"].value_counts().sort_index(ascending=False)
```

```python
#created four groups: "sub_40000", "40000_90000", "90000_125000","above_125000"
autos.loc[autos.loc[:,"odometer_km"].between(0,40000),"odom_group"]="sub_40000"
autos.loc[autos.loc[:,"odometer_km"].between(40000,90000),"odom_group"]="40000_90000"
autos.loc[autos.loc[:,"odometer_km"].between(90000,125000),"odom_group"]="90000_125000"
autos.loc[autos.loc[:,"odometer_km"]>125000,"odom_group"]="above_125000"
```

```python
avg_by_group={}
for car in autos["odom_group"].unique():
    selected_rows=autos[autos["odom_group"]==car]
    new_avg=selected_rows["price"].mean(axis=0)
    avg_by_group[car]=new_avg.round(2)

avg_by_group
```

In creating a few odometer categories, I can observe that there is a negative correlation between odometer reading and avg vehicle price. 

```python
avg_by_damage={}
for car in autos["unrepaired_damage"].unique():
    selected_rows=autos[autos["unrepaired_damage"]==car]
    avg_price=selected_rows["price"].mean(axis=0)
    avg_by_damage[car]=avg_price

avg_by_damage
```

Observation: In comparing vehicles with unrepaired damage and with repaired damage, I can conclude that vehicles for sale which do not have any needed repairs are about 3x more expensive than those with necessary repairs.

```python
#created a function that creates a dictionary aggregated average as the value with a category as the key
def new_dict(input1,input2):
    avg_dict={}
    for car in autos[input1].unique():
        selected_rows=autos[autos[input1]==car]
        avg_value=selected_rows[input2].mean(axis=0)
        avg_dict[car]=avg_value
    return avg_dict
```
